### Linear Response Theory

```python
def pLIFCond0(Re, Rp, Rs, Rv, Rex, be, bp, bs, bv, bex, mu, Cm, gL, Vre, EL, Ee, Ei, Nv, Vlb, Vth):
    # Create mesh values with linear spacing
    vs = np.linspace(Vlb, Vth, Nv)

    # Initialize variables
#     p0 = 0
    p0last = 0
    p0int = 0
    v0 = 0
    p0 = np.zeros(Nv)

    tauL = Cm / gL
    EPrime = (EL + tauL*Re*(be+be*be)*Ee + tauL*Rex*(bex+bex*bex)*Ee + tauL*Rp*(bp+bp*bp)*Ei + tauL*Rs*(bs+bs*bs)*Ei + tauL*Rv*(bv+bv*bv)*Ei) / (1 + tauL*(Re*(be+be*be) + Rp*(bp+bp*bp) + Rs*(bs+bs*bs) + Rv*(bv+bv*bv)))
    gLPrime = gL * (1 + tauL*(Re*(be+be*be) + Rex*(bex+bex*bex) + Rp*(bp+bp*bp) + Rs*(bs+bs*bs) + Rv*(bv+bv*bv)))

    # Iterate backward through mesh
    for k in range(Nv - 2, -1, -1):
        vleft = vs[k]  # Left endpoint of mesh
        vright = vs[k+1]  # Right endpoint of mesh
        vmid = (vleft + vright) / 2  # Potential at midpoint of mesh
        dv = vright - vleft  # Current mesh step

        # Variance at left, mid, and right points
        Dleft = 0.5 * (Re*be**2 * (vleft - Ee)**2 + Rex*bex**2 * (vleft - Ee)**2 + Rp*bp**2 * (vleft - Ei)**2 + Rs*bs**2 * (vleft - Ei)**2 + Rv*bv**2 * (vleft - Ei)**2)
        Dmid = 0.5 * (Re*be**2 * (vmid - Ee)**2 + Rex*bex**2 * (vmid - Ee)**2 + Rp*bp**2 * (vmid - Ei)**2 + Rs*bs**2 * (vmid - Ei)**2 + Rv*bv**2 * (vmid - Ei)**2)
        Dright = 0.5 * (Re*be**2 * (vright - Ee)**2 + Rex*bex**2 * (vright - Ee)**2 + Rp*bp**2 * (vright - Ei)**2 + Rs*bs**2 * (vright - Ei)**2 + Rv*bv**2 * (vright - Ei)**2)

#         print("Dmid = ",Dmid)
        # Drift and variance terms at left, mid, and right points
        Gleft = -(-gLPrime*(vleft - EPrime) + mu) / (Dleft * Cm)
        Gmid = -(-gLPrime*(vmid - EPrime) + mu) / (Dmid * Cm)
        Gright = -(-gLPrime*(vright - EPrime) + mu) / (Dright * Cm)

        # Simpson's rule integration for Gint2 (from left to midpoint) and Gint (over the full interval)
        vmidmid = (vleft + vmid) / 2
        Dmidmid = 0.5 * (Re*be**2 * (vmidmid - Ee)**2 + Rex*bex**2 * (vmidmid - Ee)**2 + Rp*bp**2 * (vmidmid - Ei)**2 + Rs*bs**2 * (vmidmid - Ei)**2 + Rv*bv**2 * (vmidmid - Ei)**2)
        Gmidmid = -(-gLPrime*(vmidmid - EPrime) + mu) / (Dmidmid * Cm)

        Gint2 = (1.0 / 6.0) * (Gleft + 4 * Gmidmid + Gmid) * (vmid - vleft)
        Gint = (1.0 / 6.0) * (Gleft + 4 * Gmid + Gright) * dv

        # Simpson's rule to integrate H * exp(Int G)
        H = (vmid > Vre) * (1 / Dmid)
        HexpIntG = (1.0 / 6.0) * H * (1 + 4 * np.exp(Gint2) + np.exp(Gint)) * dv

        # Update p0
        p0[k] = p0[k+1] * np.exp(Gint) + HexpIntG

        # Trapezoidal approximation to integral
        p0int += dv * ((p0[k] + p0[k+1]) / 2.0)
        v0 += dv * ((p0[k] * vleft + p0last * vright) / 2.0)
        p0last = p0[k]

    r0 = 1 / p0int  # rate is 1/(integral of p0)
    v0 = v0 * r0  # membrane potential is (integral of p0*v)*rate
    
    p0 = p0 / p0int

    return r0, v0, p0


def pLIFCond1(Re, Rp, Rs, Rv, Rex, be, bp, bs, bv, bex, mu, Cm, gL, VT, Vre, EL, Ee, Ei, Nv, Vlb, Vth, freqs, p0):
    vs = np.linspace(Vlb, Vth, Nv)
    numfreqs = len(freqs)

    # Initialize complex-valued output arrays
    Sr = np.zeros(numfreqs, dtype=complex)
    Sv = np.zeros(numfreqs, dtype=complex)
    
    tauL = Cm / gL
    EPrime = (EL + tauL*Re*(be+be*be)*Ee + tauL*Rex*(bex+bex*bex)*Ee + tauL*Rp*(bp+bp*bp)*Ei + tauL*Rs*(bs+bs*bs)*Ei + tauL*Rv*(bv+bv*bv)*Ei) / (1 + tauL*(Re*(be+be*be) + Rp*(bp+bp*bp) + Rs*(bs+bs*bs) + Rv*(bv+bv*bv)))
    gLPrime = gL * (1 + tauL*(Re*(be+be*be) + Rex*(bex+bex*bex) + Rp*(bp+bp*bp) + Rs*(bs+bs*bs) + Rv*(bv+bv*bv)))

    # Iterate over all frequencies
    for i in range(numfreqs):
        p1 = 0 + 0j
        pr = 0 + 0j
        p1last = p1
        prlast = pr
        j1 = 0 + 0j
        jr = 1 + 0j
        p1vint = 0 + 0j
        prvint = 0 + 0j
        flag = False

        for k in range(Nv - 2, -1, -1):
            dv = vs[k+1] - vs[k]
            if dv <= 0:
                raise ValueError("dv negative or too small.")
                
            

            # Evaluate G, H1, and Hr at the left endpoint
            v = vs[k]
            p00 = p0[k]
            A0 = (-gLPrime * (v - EPrime) + mu) / Cm
            A1 = 1 / Cm
            D = 0.5 * (Re*be**2 * (v - Ee)**2 + Rex*bex**2 * (v - Ee)**2 + Rp*bp**2 * (v - Ei)**2 + Rs*bs**2 * (v - Ei)**2 + Rv*bv**2 * (v - Ei)**2)
            Gleft = -A0 / D
            H1left = (j1 - A1 * p00) / D
            Hrleft = jr / D

            # Evaluate G, H1, and Hr at the midpoint
            v = (vs[k] + vs[k+1]) / 2.0
            p00 = (p0[k] + p0[k+1]) / 2.0
            A0 = (-gLPrime * (v - EPrime) + mu) / Cm
            A1 = 1 / Cm
            D = 0.5 * (Re*be**2 * (v - Ee)**2 + Rex*bex**2 * (v - Ee)**2 + Rp*bp**2 * (v - Ei)**2 + Rs*bs**2 * (v - Ei)**2 + Rv*bv**2 * (v - Ei)**2)
            Gmid = -A0 / D
            H1mid = (j1 - A1 * p00) / D
            Hrmid = jr / D

            # Evaluate G, H1, and Hr at the right endpoint
            v = vs[k+1]
            p00 = p0[k+1]
            A0 = (-gLPrime * (v - EPrime) + mu) / Cm
            A1 = 1 / Cm
            D = 0.5 * (Re*be**2 * (v - Ee)**2 + Rex*bex**2 * (v - Ee)**2 + Rp*bp**2 * (v - Ei)**2 + Rs*bs**2 * (v - Ei)**2 + Rv*bv**2 * (v - Ei)**2)
            Gright = -A0 / D
            H1right = (j1 - A1 * p00) / D
            Hrright = jr / D

            # Simpson's approx to exponential of integral of G
            expGint = np.exp((1.0 / 6.0) * (Gleft + 4 * Gmid + Gright) * dv)

            # Evaluate G halfway between left endpoint and midpoint
            v = (vs[k+1] + 3.0 * vs[k]) / 4.0
            p00 = (p0[k+1] + 3.0 * p0[k]) / 4.0
            A0 = (-gLPrime * (v - EPrime) + mu) / Cm
            D = 0.5 * (Re*be**2 * (v - Ee)**2 + Rex*bex**2 * (v - Ee)**2 + Rp*bp**2 * (v - Ei)**2 + Rs*bs**2 * (v - Ei)**2 + Rv*bv**2 * (v - Ei)**2)
            Gmidmid = -A0 / D

            # Estimate the integral of G from the left endpoint to the midpoint
            expGint2 = np.exp((1.0 / 6.0) * (Gleft + 4.0 * Gmidmid + Gmid) * (dv / 2.0))

            # Estimate the integral of H*exp(Int G) over the entire interval
            H1expIntG = (1.0 / 6.0) * (H1left + 4.0 * H1mid * expGint2 + H1right * expGint) * dv
            HrexpIntG = (1.0 / 6.0) * (Hrleft + 4.0 * Hrmid * expGint2 + Hrright * expGint) * dv

            p1 = p1 * expGint + H1expIntG
            pr = pr * expGint + HrexpIntG

            j1 += 2 * np.pi * 1j * freqs[i] * p1 * dv
            jr += 2 * np.pi * 1j * freqs[i] * pr * dv

            if not flag and v <= Vre:
                jr -= 1
                flag = True

            # Trapezoidal approximation to integral
            p1vint += ((p1 + p1last) / 2) * v * dv
            prvint += ((pr + prlast) / 2) * v * dv

            p1last = p1
            prlast = pr
            
#             print(f'freq={freqs[i]} k={k} p1={p1} pr={pr} j1={j1} jr={jr}')

        Sr[i] = -j1 / jr
        Sv[i] = p1vint + Sr[i] * prvint

    return Sr, Sv
```
